```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms

import json
import cv2
import numpy as np
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
class KeypointsDataset(Dataset):
    def __init__(self, img_dir, data_file):
        self.img_dir = img_dir
        with open(data_file, "r") as f:
            self.data = json.load(f)

        self.transforms = transforms.Compose([
            transforms.ToPILImage(),
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        img = cv2.imread(f"{self.img_dir}/{item['id']}.png")
        h,w = img.shape[:2]

        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = self.transforms(img)
        kps = np.array(item['kps']).flatten()
        kps = kps.astype(np.float32)

        kps[::2] *= 224.0 / w # Adjust x coordinates
        kps[1::2] *= 224.0 / h # Adjust y coordinates

        return img, kps
```

Langkah 3: Buat dan muat data pelatihan dan validasi

```python
train_dataset = KeypointsDataset("data/images","data/data_train.json")
val_dataset = KeypointsDataset("data/images","data/data_val.json")

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=True)
```

Step 4: Load pre-trained Resnet50 model and mmodifikasi bagian Fully Connected (fc) Layer dengan mengganti lapisan terakhir yang terhubung sepenuhnya dengan lapisan baru yang menghasilkan 28 nilai yang merepresentasikan 14 keypoints, masing-masing dengan koordinat x dan y (karenanya 2 nilai per titik kunci).

Mengapa Memodifikasi Lapisan Terakhir?
Model ResNet-50 asli dirancang untuk klasifikasi (1000 kelas), tetapi tugas ini melibatkan deteksi keypoint, yang tujuannya adalah untuk memprediksi koordinat (x, y) keypoint. Dengan mengganti lapisan yang terhubung sepenuhnya, model dapat disesuaikan untuk menghasilkan 28 nilai yang diperlukan (14 titik kunci × 2).

```python
model = models.resnet50(pretrained=True)
model.fc =  torch.nn.Linear(model.fc.in_features, 14*2) # Replaces the last layer
```

```python
model = model.to(device)
```

Step 5: Definisikan Loss Function dan Optimizer

```python
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

Step 6: Train Model

Epoch Loop: Melakukan iterasi sebanyak epochs (20 kali).

Batch Loop: Pada setiap epoch, iterasi melalui batch-batch data dari train_loader.

Training Step:
- Reset gradien (optimizer.zero_grad()).
- Lakukan forward propagation untuk mendapatkan prediksi.
- Hitung loss antara prediksi dan ground truth.
- Lakukan backward propagation untuk menghitung gradien.
- Update parameter model (optimizer.step()).

Logging: Setiap 10 batch, mencetak nilai loss untuk memantau proses training.

```python
epochs=20
for epoch in range(epochs):
    for i, (imgs,kps) in enumerate(train_loader):
        imgs = imgs.to(device)
        kps = kps.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, kps)
        loss.backward()
        optimizer.step()

        if i % 10 == 0:
            print(f"Epoch {epoch}, iter {i}, loss: {loss.item()}")
```

Step 7: Save Model

```python
torch.save(model.state_dict(), "keypoints_model.pth")
```
